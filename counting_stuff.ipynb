```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread("Preprocessed dataset/IMG_Niguarda1_936_129_F6_T975.png", cv.IMREAD_GRAYSCALE)
```

```python
ret, labels = cv.connectedComponents(img)
print(ret-1)
```

```python
plt.subplot(221)
plt.title('Default img')
plt.imshow(img, cmap="gray", vmin=0, vmax=255)
```

```python
kernel = np.ones((5,5), np.uint8)
img_dilation = cv.dilate(img, kernel, iterations=1)
img_erode = cv.erode(img_dilation,kernel, iterations=1)
#img_erode = cv.medianBlur(img_erode, 7)
```

```python

plt.subplot(221)
plt.title('Dilatation + erosion')
plt.imshow(img_erode, cmap="gray", vmin=0, vmax=255)
```
